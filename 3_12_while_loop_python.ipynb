```python
#Write a python program to find the multiplication of N natural numbers.

Number = int(input('Enter a number: '))
m_total = 1
i = 1
while i<= Number:
    m_total *= i
    i+=1

print(f'multiplication of 1 to {Number} Natural number is: {m_total}')
```

```python
#Iterate in list Using while loop

days = 0    
week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'] #list
while days < 7:
   print("Today is " + week[days])
   days += 1   #day = day+1
```

```python
#Iterate in tuple Using while loop

my_tuple = ('premanand', 35,'CPP:Python')

i = 0

while i < len(my_tuple):
    print(my_tuple[i])
    i+=1
```

```python
#Iterate in string Using while loop

my_string = 'PREMANANDS'

j = 0
while j<len(my_string):
    print(my_string[j])
    j+=1
```

```python
# Program to add natural
# numbers up to 
# sum = 1+2+3+...+n

# To take input from the user,
n = int(input("Enter n: "))

#n = 10

# initialize sum and counter
sum = 0
i = 1

while i <= n:
    sum = sum + i
    i = i+1    # update counter

# print the sum
print("The sum is", sum)
```

```python
#Illustrate the use of else statement with the while loop
counter = 0

while counter < 3:
    print("Inside loop")
    counter = counter + 1
else:
    print("Inside else")
```
